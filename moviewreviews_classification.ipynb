```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
```

```python
ip = df['review']
op = df['label']
```

```python
tfidf = TfidfVectorizer(max_features=5000)
ip_tfidf = tfidf.fit_transform(ip)
```

```python
ip_train, ip_test, op_train, op_test = train_test_split(ip_tfidf, op, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
model.fit(ip_train, op_train)
```

```python
op_pred = model.predict(ip_test)
```

```python
print("Accuracy:", accuracy_score(op_test, op_pred))
print(classification_report(op_test, op_pred))
```
